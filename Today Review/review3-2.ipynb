### 비대칭도(왜도)

- 왜도는 실수 값 확률 변수의 확률 분포 비대칭성을 나타내는 지표
- 왜도의 값이 양수일 때 오른쪽 부분에 긴꼬리
- 왜도의 값이 음수일 때 왼쪽 부분에 긴꼬리
- 평균 = 중앙값 : 왜도는 0!

```python
df.skew()
#음수면 왼쪽 긴꼬리
```

### 첨도

- 첨도는 확률 분포의 뾰족한 정도를 나타내는 척도
- 관측치 값들이 어느 정도 집중적으로 중심에 몰려 있는가를 측정할 때 사용
- 첨도 기본값:3
- 첨도 값이 3 이상: 뾰족
- 첨도 값이 3 이하: 완만

```python
df.kurt()
```

#### 상관분석

- r = +1 : 완전히 동일
- r = -1 : 전혀 다르다
- r= 0 : 무관

```python
corr=df.corr()
corr
```

```python
mask=np.triu(np.ones_like(corr)) 
#np.ones_like(x) : x와 크기만 같은 1로 이루어진 array를 생성
mask
```

```python
# heatmap 을 통해 상관계수를 시각화
sns.heatmap(corr,annot=True,cmap='RdBu',vmax=1,vmin=-1,mask=mask)
```
