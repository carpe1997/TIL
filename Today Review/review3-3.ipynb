## 범주형 변수

### 범주형 데이터 유일값의 빈도수

```python
df.nunique()
```

```python
sns.countplot(data=df,x="origin")
```

### 1개 변수의 빈도수

```python
df["origin"].value_counts()
```

### 2개 이상의 변수에 대한 빈도수

```python
sns.countplot(data=df,x="origin",hue="cylinders")
```

```python
sns.countplot(data=df,x="cylinders",hue="origin")
```

```python
pd.crosstab(df["origin"],df["cylinders"],normalize=True)
```

### 범주형 vs 수치형변수

```python
sns.barplot(data=df,x="origin",y="mpg",ci=None)
```

### groupby를 통한 연산

```python
df.groupby(["origin"])["mpg"].mean()
```

### pivot table을 통한 연산

- index 와 values가 필요

```python
pd.pivot_table(data=df,index="origin",values="mpg") #평균값
```

```python
sns.barplot(data=df,x="origin",y="mpg",estimator=np.sum,ci=None)
```

```python
sns.barplot(data=df,x="cylinders",y="mpg",hue="origin",ci=None)
```

```python
df.groupby(["cylinders","origin"])["mpg"].mean().unstack()
```

```python
pd.pivot_table(data=df,index="cylinders",columns="origin",values="mpg")
```

```python
sns.barplot(data=df,x="origin",y="mpg",hue="cylinders",ci=None)
```

### boxplot과 사분위수

```python
sns.boxplot(data=df,x="origin",y="mpg")
```

#### 이상값을 알고싶다면


```python
g=df.groupby("origin")["mpg"].describe()
g
```

```python
g.loc["europe","75%"]
```

```python
Q3=g.loc["europe","75%"]
Q1=g.loc["europe","25%"]
IQR=Q3-Q1
out_max=Q3+1.5*IQR
out_min=Q1-1.5*IQR
```

```python
out_max
```

```python
out_min
```

```python
europe=df[df["origin"]=="europe"]
sns.boxplot(data=europe,x="origin",y="mpg")
```

```python
sns.boxenplot(data=europe,x="origin",y="mpg")
```

```python
sns.violinplot(data=europe,x="mpg",y="origin")
```

### catplot을 통한 범주형 데이터의 서브플롯 시각화

```python
sns.catplot(data=df,x="origin",y="mpg",col="cylinders",col_wrap=3,kind="bar",ci=None)
```

```python
sns.catplot(data=df,x="origin",y="mpg",col="cylinders",col_wrap=3,kind="violin")
```

```python
sns.catplot(data=df,x="origin",col="cylinders",col_wrap=3,kind="count")
```
