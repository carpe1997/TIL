```python
def movie_review(page_no):
    url=f"https://movie.naver.com/movie/point/af/list.nhn?&page={page_no}"
    response=requests.get(url)
    html=bs(response.text,"lxml")
    temp=html.select("table")
    table=pd.read_html(str(temp))
    movie_review=table[0].dropna()
    return movie_review
```

```python
movie_review(4)
```

```python
movie_review(5)
```

```python
review=[]

for page_no in range(1,10):
    k=movie_review(page_no)
    review.append(k)
len(review)
```

```python
df=pd.concat(review)
df.shape
```
